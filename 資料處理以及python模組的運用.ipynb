# **list的處理**

```python
lst_1 = []
for i in range(1, 21):
  lst_1.append(i)
lst_2 = []
for i in lst_1:
  temp = i
  if(i % 2 == 0):
    temp *= 2
    lst_2.append(temp)
  else:
    temp += 1
    lst_2.append(temp)
lst_2
```

# **np.array的加減運算**

```python
import numpy as np
data = np.array([[10, 20, 30],
          [40, 50, 60],
          [70, 80, 90]])
data[data > 50] = 0
print(f"data:\n{data}")
print()
sum_0 = np.sum(data, axis = 0)
sum_1 = np.sum(data, axis = 1)
print(f"沿著axis0相加:{sum_0}")
print(f"沿著axis1相加:{sum_1}")
```

# **利用yahoo股市的資料，將台積電(2330)的交易變化量用圖表表示**

```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
_2330 = yf.download("2330.TW", period = "1y")
_2330["Value"] = _2330["Close"] * _2330["Volume"]
_2330["Value_Change"] = _2330["Value"].pct_change()
plt.plot(_2330["Value"], label = "2330.TW Value Change Ratio")
plt.title("2330.TW Trading Change Ratio")
plt.ylabel("Change Ratio")
plt.legend()
plt.show()
```

# **利用yahoo股市計算apple, google, microsoft和元大50的年化預期收益、共變異數矩陣、標準差、投資組合**

```python
import numpy as np
import pandas as pd
import yfinance as yf

stocks = ["AAPL", "GOOGL", "MSFT", "0050.TW"]
s = "2024-01-01"
e = "2025-06-04"
df = yf.download(stocks, start = s, end = e)["Close"]
returns = df.pct_change().dropna()
expected_returns = returns.mean() * 252
covariance_matrix = returns.cov() * 252
std_returns = returns.std() * 252
```

```python
print(f"\n預期回報(年化)\n{expected_returns}")
```

```python
print("\n 共變異數矩陣 ( 年化 ):\n", covariance_matrix)
```

```python
weights = np.array([0.4, 0.3, 0.2, 0.1])

portfolio_expected_return = np.dot(weights, expected_returns)
print("\n 投資組合預期回報 ( 年化 ):", portfolio_expected_return)

portfolio_variance = np.dot(weights.T, np.dot(covariance_matrix, weights))
print(" 投資組合變異數 ( 年化 ):", portfolio_variance)


```

```python
portfolio_STD = np.sqrt(portfolio_variance)
print(" 投資組合標準差 ( 年化 ):", portfolio_STD)
```
